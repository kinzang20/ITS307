**Q4.** The annual salaries of employees in a large company are approximately normally distributed with a mean of $\$50,000$ and a standard deviation of $\$20,000$.

**a)** What percent of people earn less than $\$40,000?$

**b)** What percent of people earn between $\$45,000$ and $\$65,000?$

**c)** What percent of people earn more than $\$70,000?$

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = np.arange(1,100000)
pdf = norm.pdf(data, loc = 50000, scale = 20000 ) 
sns.set_style('whitegrid')
sns.lineplot(x = data, y = pdf , color = 'black')
plt.xlabel('Annual Salaries')
plt.ylabel('Probability Density')
plt.show()
```

```python
prob_1 = norm(loc = 50000 , scale = 20000).cdf(40000)
print((prob_1*100),'% earns less than $40,000' )
 
cdf_upper_limit = norm(loc = 50000 , scale = 20000).cdf(65000)
cdf_lower_limit = norm(loc = 50000 , scale = 20000).cdf(40000)
 
prob_2 = cdf_upper_limit - cdf_lower_limit
print(prob_2*100, '% earns between $40,000 and $65,000')
prob_3 = norm(loc = 50000 , scale = 20000).cdf(70000)
print((1-prob_3)*100, "% earns more than $70,000")
```

```python
import numpy as np
x = [69,63,66,64,67,64,70,66,68,67,65,71]
y = [70,65,68,65,69,66,68,65,71,67,64,72]
deg = 1
p = np.polyfit(x, y, deg)
b = p[0]  
a = p[1]  
print(f'The fitted straight line has equation y = {b:.1f}x {a:=+6.1f}')
```

```python
import pandas as pd
data = {'weight of father(X)': [69,63,66,64,67,64,70,66,68,67,65,71], 'Weight of son(Y)': [70,65,68,65,69,66,68,65,71,67,64,72]}  
df = pd.DataFrame(data)
df
```

```python
y_pred = [[89],[83],[85],[62],[58],[77],[80],[90],[62],[99],[100],[57]]
lr.predict(y_pred)
```
