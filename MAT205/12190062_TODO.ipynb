Find the variance and expectation of the exponential distribution.

```python
import random as rand
s, s2 = 0, 0
num_trials = int(1e5)

for i in range(num_trials):
    x = rand.random()
    s += x
    s2 += x*x

mean = s/num_trials
mean_x2 = s2/num_trials

var = mean_x2 - mean * mean

print("Estimated mean/Expectation = {0:.4f}, variance = {1:.4f}".format(mean, var))
```
