**Q4.** The annual salaries of employees in a large company are approximately normally distributed with a mean of $\$50,000$ and a standard deviation of $\$20,000$.



**Solution:**

A Standard Normal variable, usually denoted by Z, can be obtained from a non-standard Normal($\mu$, $\sigma$) random variable X by standardizing, that is, subtracting the mean and dividing by the standard deviation,

$Z = \frac{X-\mu}{\sigma}$

Given; 

$\mu$ = $\$50,000$ 

$\sigma$ = $\$20,000$

Let X be the normal variable denoting the annual salaries of employees.

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

mu = 50000
sigma = 20000 
# Creating the distribution
data = np.arange(0, 100000, 10)
pdf = norm.pdf(data, mu, sigma ) 
 
#Visualizing the distribution
sns.set_style('whitegrid')
sns.lineplot(x = data, y = pdf , color = 'black')
plt.xlabel('salaries')
plt.ylabel('Probability Density')
plt.show()
```

**a)** What percent of people earn less than $\$40,000?$



Given;

$P (X < 40,000) = P(X < 40,000)$ 

$Z = \frac{X-\mu}{\sigma}$

$P (X < 40,000) = P(Z < \frac{X-\mu}{\sigma})$ 

$P (X < 40,000) = P(Z < \frac{40000 - 50000}{20000})$ 

$P (X < 40,000) = P(Z < -0.5)$ 

By using symmetry rule 

$P (X < 40,000) = P(0.5 < Z)$

$P (X < 40,000) = 0.5 - P(0 < Z < 0.5)$

$P (X < 40,000) = 0.5 - 0.1915$

$P (X < 40,000) = 0.3085$

Therefore, people who earn less than $40,000 is 0.3085  100 = **30.85%**



```python
#Probability of salaries less than $40,000.
prob_1 = norm(mu , sigma).cdf(40000)
print(prob_1)
```

**b)** What percent of people earn between $\$45,000$ and $\$65,000?$



Given ;

$P(45,000 < X < 65,000)$

$Z = \frac{X-\mu}{\sigma}$

$P (45,000 < X < 65,000) = P(\frac{X-\mu}{\sigma} < Z < \frac{X-\mu}{\sigma})$ 

$P (45,000 < X < 65,000) = P(\frac{45000 - 50000}{20000} < Z < \frac{65000-50000}{20000})$ 

$P (45,000 < X < 65,000) = P(-0.25 < Z < 0.75)$ 

$P (45,000 < X < 65,000) = P(-0.25 < Z < 0) + P(0 < Z < 0)$ 

$P (45,000 < X < 65,000) = 0.0987 + 0.2734$ 

$P (45,000 < X < 65,000) = 0.3721$ 

Therefore, the percent of people who earn between $\$45,000$ and $\$65,000$ is 0.3721  100 = **37.21%**

```python
#probability that the salaries of people who earn between $45,000 and $65,000. 
cdf_upper_limit = norm(mu , sigma).cdf(65000)
cdf_lower_limit = norm(mu , sigma).cdf(45000)
 
prob_2 = cdf_upper_limit - cdf_lower_limit
print(prob_2)
```
